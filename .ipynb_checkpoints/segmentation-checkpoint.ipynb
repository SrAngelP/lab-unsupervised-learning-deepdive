```python
pca=PCA(n_components = 1)
data_pca = pca.fit_transform(data_scale)
```

```python
#Your code here:

dbscan=DBSCAN(eps=0.7, min_samples=10).fit(data_pca)
km_fit=KMeans(n_clusters=2)
km_fit.fit(data_pca)
```

```python
data['labels'] = km_fit.labels_
data['labels_dbscan'] = dbscan.labels_
```

```python
plt.scatter(data['age'], data['friends'], c = data['labels'])
```

```python
plt.scatter(data['age'], data['friends'], c = data['labels_dbscan'])
```
